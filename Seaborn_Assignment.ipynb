```python
ANS-1

Seaborn is a powerful data visualization library based on Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Some of the plots that can be created using Seaborn along with their uses are:

Scatter Plot:
Use: Visualize the relationship between two continuous variables to identify correlations or patterns in the data.

Bar Plot:
Use: Compare categorical data by displaying rectangular bars of different heights or lengths.

Histogram:
Use: Visualize the distribution of a continuous variable and understand its frequency distribution.

Box Plot:
Use: Display the distribution of data and identify potential outliers, quartiles, and the median.

Heatmap:
Use: Display a 2D representation of data using color intensities to show the magnitude of values in a matrix.




ANS-2


To plot a line plot using the "fmri" dataset from Seaborn, we can follow these steps:

Import the necessary libraries.
Load the "fmri" dataset using seaborn.load_dataset().
Create the line plot using seaborn.lineplot() with "timepoint" on the x-axis and "signal" on the y-axis, while distinguishing the lines based on the "event" and "region" columns.
Here's the code to achieve this:


import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create the line plot
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")
plt.title("Line Plot of Signal vs. Timepoint for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()




ANS-3


Import the necessary libraries.
Load the "titanic" dataset using seaborn.load_dataset().
Create the first box plot using seaborn.boxplot() with 'pclass' on the x-axis and 'age' on the y-axis.
Create the second box plot using seaborn.boxplot() with 'pclass' on the x-axis and 'fare' on the y-axis.
Here's the code to achieve this:


import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot (age vs. pclass)
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title("Box Plot of Age vs. Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

# Create the second box plot (fare vs. pclass)
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title("Box Plot of Fare vs. Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

# Display the plots
plt.show()




ANS-4


```
